```python
def list_operations():
    my_list = []
    while True:
        print("\nList operations:")
        print("1. Add element")
        print("2. Remove element")
        print("3. Modify element")
        print("4. View list")
        print("5. Exit")
        choice = input("Choose an operation: ")

        if choice == '1':
            element = input("Enter element to add: ")
            my_list.append(element)
            print("Element added.")
        elif choice == '2':
            element = input("Enter element to remove: ")
            if element in my_list:
                my_list.remove(element)
                print("Element removed.")
            else:
                print("Element not found.")
        elif choice == '3':
            index = int(input("Enter index of element to modify: "))
            if 0 <= index < len(my_list):
                new_element = input("Enter new element: ")
                my_list[index] = new_element
                print("Element modified.")
            else:
                print("Invalid index.")
        elif choice == '4':
            print("Current list:", my_list)
        elif choice == '5':
            break
        else:
            print("Invalid choice.")

def dict_operations():
    my_dict = {}
    while True:
        print("\nDictionary operations:")
        print("1. Add element")
        print("2. Remove element")
        print("3. Modify element")
        print("4. View dictionary")
        print("5. Exit")
        choice = input("Choose an operation: ")

        if choice == '1':
            key = input("Enter key to add: ")
            value = input("Enter value: ")
            my_dict[key] = value
            print("Element added.")
        elif choice == '2':
            key = input("Enter key to remove: ")
            if key in my_dict:
                del my_dict[key]
                print("Element removed.")
            else:
                print("Key not found.")
        elif choice == '3':
            key = input("Enter key to modify: ")
            if key in my_dict:
                new_value = input("Enter new value: ")
                my_dict[key] = new_value
                print("Element modified.")
            else:
                print("Key not found.")
        elif choice == '4':
            print("Current dictionary:", my_dict)
        elif choice == '5':
            break
        else:
            print("Invalid choice.")

def set_operations():
    my_set = set()
    while True:
        print("\nSet operations:")
        print("1. Add element")
        print("2. Remove element")
        print("3. Modify element (remove and add)")
        print("4. View set")
        print("5. Exit")
        choice = input("Choose an operation: ")

        if choice == '1':
            element = input("Enter element to add: ")
            my_set.add(element)
            print("Element added.")
        elif choice == '2':
            element = input("Enter element to remove: ")
            if element in my_set:
                my_set.remove(element)
                print("Element removed.")
            else:
                print("Element not found.")
        elif choice == '3':
            element = input("Enter element to remove: ")
            new_element = input("Enter element to add: ")
            if element in my_set:
                my_set.remove(element)
                my_set.add(new_element)
                print("Element modified.")
            else:
                print("Element not found.")
        elif choice == '4':
            print("Current set:", my_set)
        elif choice == '5':
            break
        else:
            print("Invalid choice.")

def main():
    while True:
        print("\nChoose a data structure to operate on:")
        print("1. List")
        print("2. Dictionary")
        print("3. Set")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            list_operations()
        elif choice == '2':
            dict_operations()
        elif choice == '3':
            set_operations()
        elif choice == '4':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
