```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('/content/householdtask3.csv')
# Scatter Plot: Income vs Expenditure
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='income', y='expenditure')
plt.title('Scatter Plot: Income vs Expenditure')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.show()

# Line Chart: Year vs Income
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='year', y='income')
plt.title('Line Chart: Year vs Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()

# Bar Chart: Average Income by Year
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='year', y='income')
plt.title('Bar Chart: Average Income by Year')
plt.xlabel('Year')
plt.ylabel('Average Income')
plt.show()

# Histogram: Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(data['income'], kde=True)
plt.title('Histogram: Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
