```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the CSV file into a DataFrame
file_path = '/content/Data Cleaning and Preprocessing.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
print("First few rows of the DataFrame:")
print(df.head())

# Display the summary of the DataFrame to understand the data types and missing values
print("\nSummary of the DataFrame:")
print(df.info())

# Handling missing values
# Fill missing values with the mean of each numeric column
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.mean()), axis=0)

# Removing outliers using IQR
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filter the data to remove outliers
df = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Filtering data based on a condition (example: filtering rows where Y-Kappa > 20)
filtered_df = df[df['Y-Kappa'] > 20]

# Dropping the 'Observation' column
filtered_df.drop(['Observation'], axis=1, inplace=True)

# Display the columns of the cleaned DataFrame
print("\nColumns of the cleaned DataFrame:")
print(filtered_df.columns)

# Calculating summary statistics of the cleaned DataFrame
summary_stats = filtered_df.describe()

# Display the summary statistics
print("\nSummary Statistics of the cleaned DataFrame:")
print(summary_stats)
```
